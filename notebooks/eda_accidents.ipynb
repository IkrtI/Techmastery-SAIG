# EDA — สถิติอุบัติเหตุทางถนนประเทศไทย (Dataset 1)

สำรวจชุดข้อมูล `thai_accidental_dataset.csv` เพื่อสนับสนุน narrative ของ RailGuard AI:
ความเสี่ยงบนถนน โดยเฉพาะพฤติกรรมฝ่าฝืนกฎจราจร

```python
import pandas as pd

CSV = "../dataset/1_ข้อมูลอุบัติเหตุทางรถยนต์ในประเทศไทย/thai_accidental_dataset.csv"
df = pd.read_csv(CSV)
df["time"] = pd.to_datetime(df["time"], errors="coerce")
print(df.shape)
df.head()
```

```python
print("ผู้เสียชีวิตรวม:", int(df.fatalities.sum()))
print("ผู้บาดเจ็บรวม:", int(df.injuries.sum()))
print("\nสาเหตุอันดับต้น:")
print(df.cause.value_counts().head(10))
```

```python
print("ยานพาหนะคันแรกที่ประสบเหตุ:")
print(df.first_vehicle.value_counts().head(10))
print("\nอุบัติเหตุตามชั่วโมง:")
print(df.time.dt.hour.value_counts().sort_index())
```

```python
# ความรุนแรง: อัตราเสียชีวิตต่อเหตุการณ์ แยกตามยานพาหนะ
sev = df.groupby("first_vehicle").agg(
    n=("fatalities", "size"), deaths=("fatalities", "sum")
)
sev = sev[sev.n >= 500].assign(death_rate=lambda d: d.deaths / d.n)
sev.sort_values("death_rate", ascending=False).head(10)
```

**ข้อสังเกตหลักสำหรับสไลด์**
- สาเหตุอันดับ 1 คือขับเร็วเกินกำหนด — สะท้อนพฤติกรรม *ฝ่าฝืนกฎ* แบบเดียวกับการฝ่าเครื่องกั้นทางรถไฟ
- รถจักรยานยนต์คือกลุ่มเสี่ยงหลัก — ตรงกับผลตรวจจับของ RailGuard ที่ผู้ฝ่าฝืนส่วนใหญ่เป็นมอเตอร์ไซค์
- ชั่วโมงเร่งด่วนเย็น (16-19 น.) เกิดเหตุสูงสุด — ตรงกับช่วงที่คนเร่งรีบฝ่าทางตัด
